```python
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
poverty_df = pd.read_csv('Resources/count_poverty_rate.csv')
poverty_df
```

```python
#cleaning up poverty dataset by selecting the year 2019 and renaming the first column
poverty_df = poverty_df.rename(columns={'Unnamed: 0': "County"})
new_poverty_df = poverty_df[['County','2019']]
new_poverty_df
```

```python
# creating string to connect to postgres
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/final_project"
```

```python
# connecting to postgres using engine
engine = create_engine(db_string)
```

```python
# Loading cleaned poverty rate data to postgres
new_poverty_df.to_sql(name='county_poverty_rate', con=engine, method='multi')
```
